```python
#experiment with different values and check what fits the model best without overfitting or underfitting
#least loss I coul get is around 0.5895 after changing the values of different layer I think this is coming 
# due to the fact the data given to model was very less.
#After getting minimum loss I tried some more combinations but results are not going below 0.5 loss value 
#Hence the model will not be able to accurately predict the results

import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

student_dataset = pd.read_csv("LinearRegressionDataset\Admission_Predict.csv")

X = student_dataset.iloc[:,1:8] 
Y = student_dataset.iloc[:,8:9]

x_train ,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=42)

x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
x_test = x_test.to_numpy()
y_test = y_test.to_numpy()



model = tf.keras.models.Sequential()
model.add(tf.keras.layers.InputLayer(input_shape=(7),batch_size=35))
model.add(tf.keras.layers.Dense(256,activation='relu'))
model.add(tf.keras.layers.Dropout(0.7))
model.add(tf.keras.layers.Dense(32,activation='softmax'))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(64))
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(16))
model.add(tf.keras.layers.Dense(32))
model.add(tf.keras.layers.Dropout(0.7))
model.add(tf.keras.layers.Dense(2))
model.add(tf.keras.layers.Dense(1))


model.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(from_logits=True))

model.fit(x_train,y_train,batch_size=35,epochs=30)
print(model.predict(x_test)[:6])
print(y_test[:6])

```
