```python
import numpy as np
import pandas as pd
from sklearn import neighbors
import matplotlib.pyplot as plt

n_neighbors = 15

iris_data = pd.read_csv("IRIS.csv")

X = np.array(iris_data.iloc[:,:3])   #TARGET DATA 
y = np.array(iris_data.iloc[:,4:])   #TARGET RESPONSE

clf = neighbors.KNeighborsClassifier(n_neighbors, weights="uniform")
clf.fit(X, y.ravel())

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection ="3d")
my_cmap = plt.get_cmap('hsv')

xs = X[:,0:1]
ys = X[:,1:2]
zs = X[:,2:3]


ax.scatter3D(xs.ravel(),ys.ravel(),zs.ravel(),c = (xs + ys + zs),cmap = my_cmap)
plt.plot(6.3, 2.7,4.9, marker='v', markersize=4, color="black")
plt.show()

```
